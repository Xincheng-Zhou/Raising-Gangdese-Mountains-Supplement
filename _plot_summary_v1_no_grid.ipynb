Median elevation (ST) vs. time plot for the Gangdese region.

```julia
using CSV
using Plots,DataFrames


files = readdir(".")


for file in files
    if endswith(file, ".csv")
        println(file)
    end
end

```

```julia
# read CSV files
data1 = CSV.read("data_Ref.csv", DataFrame)
data2 = CSV.read("data_R_TNIC.csv", DataFrame)
data3 = CSV.read("data_R_TKIC.csv", DataFrame)
data6 = CSV.read("data_R_Bf_80.csv", DataFrame)
data7 = CSV.read("data_R_Bf_120.csv", DataFrame)
data8 = CSV.read("data_R_Bf_no.csv", DataFrame)
data9 = CSV.read("data_R_M.csv", DataFrame)
data15 = CSV.read("data_R_ECL_12.csv", DataFrame)
data17 = CSV.read("data_R_14_6.csv", DataFrame)
data18 = CSV.read("data_R_15_8.csv", DataFrame)
data19 = CSV.read("data_R_ECL_8.csv", DataFrame)

println(names(data1))
```

```julia
p1 = plot(data2[!, :time], data2[!, :arc_ST_median],
    label="15km",
    #title="Crustal Thickness & Elevations @ t=$time",
    ylabel="elevation (km)",
    xlabel="time (Myr)", xlims=(0,20), ylims=(-1,7),
    grid=false, xticks=0:1:20, yticks=-3:1:8, size=(1200,300),linewidth=2,color=:blue,linestyle=:dash, dpi = 300, bg=:transparent, legendfontsize=10,tickfontsize=12)



plot!(p1, data1[!, :time], data1[!, :arc_ST_median], label="25km (Ref)", linewidth=2, color=:blue)
plot!(p1, data3[!, :time], data3[!, :arc_ST_median], label="30km", linewidth=2, color=:blue,linestyle=:dashdot)


display(p1)

savefig(p1, "plot_diff_indian_thickness_no_grid.png")  
```

```julia
p1 = plot(data6[!, :time], data6[!, :arc_ST_median],
    label="80km depth slab breakoff",
    #title="Crustal Thickness & Elevations @ t=$time",
    ylabel="elevation (km)",
    xlabel="time (Myr)", xlims=(0,20), ylims=(-1,7),
    grid=false, xticks=0:1:20, yticks=-3:1:8, size=(1200,300),linewidth=2,color=:blue, linestyle=:dot, dpi = 300, bg=:transparent, legendfontsize=10,tickfontsize=12)



plot!(p1, data1[!, :time], data1[!, :arc_ST_median], label="100km depth slab breakoff (Ref)",linewidth=2, color=:blue)
plot!(p1, data7[!, :time], data7[!, :arc_ST_median], label="120km depth slab breakoff", linestyle=:dashdot,linewidth=2, color=:blue)
plot!(p1, data8[!, :time], data8[!, :arc_ST_median], label="no slab breakoff", linestyle=:dash,linewidth=2, color=:blue)



display(p1)

savefig(p1, "plot_diff_breakoff_no_grid.png")  
```
